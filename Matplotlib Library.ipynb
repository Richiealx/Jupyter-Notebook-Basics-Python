# Matplotlib Library

## What is Matplotlib?

### Matplotlib is an open source drawing library which supports rich drawing types
### It is used to draw 2-D and 3-D graphics.
### You can understand your data easily by visualizing it with the help of matplotlib.
### You can generate plots,histograms,bar charts, and many other charts with just few lines of code

## Types of Plots
### 1.Bar Chart
### 2.Scatter Plot
### 3.Histogram
### 4.Line Chart
### 5.Pie Chart
### 6.Areas Graph

### What is pylab?

```python
# Pylab is a module within Matplotlib that was designed to simplify the process of plotting 
# by combining the functionality of NumPy and Matplotlib into a single namespace. 
# This allowed users to import pylab and have access to both plotting functions (from Matplotlib) 
# and array operations (from NumPy) without needing to import each module separately.
```

```python
# Print the pylab version

from matplotlib import pylab

print(pylab.__version__)
```

```python
# Modern Approach:
# Instead of using pylab, the recommended approach today is to import Matplotlib and NumPy separately:

import matplotlib.pyplot as plt

import numpy as np
```

```python
# Use numpy to generate random data

import numpy as np

x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
```

```python
# reshape x,y array

print(np.array([x,y]).reshape(25,2))
```

```python
# Reshape (25,2) to (2,25) where x=2, and y=25

print(np.array([x,y]).reshape(2,25))
```

```python
# NOTE: Matplotlib works with array of x and array of y

"""It only takes 1 command line to draw the data"""

# Plot a line of x and y array

pylab.plot(x,y,'r') #where 'r' stands for red, 'b' stands for blue, 'y' stands for yellow, 'g' stands for green
```

```python
# Plot a line of x and y array
# Make the line blue

pylab.plot(x,y,'b') #where 'b' is blue
```

```python
# Plot a line of x and y array
# Make the line green

pylab.plot(x,y,'g') #where 'g' is green
```

```python
"""NOTE: Deciding on what color to use to plot a line is very important because, it enables stakeholders 
for example to determine that red color can represent losses, while green color represent profits in their company"""
```

```python
# Drawing a subgraph

pylab.subplot(1,2,1) # The contents of the brackets represents rows, columns and index. rows=1, columns=2, index=1

pylab.plot(x,y,'r--') # where 'r' represents color red, and (break red color--) is line style.

pylab.subplot(1,2,2)
pylab.plot(y,x,'*g-')
```

```python
# We can also change the columns and row numbers as well as indexes
# change subplot 2 to (2,2,2)

pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(2,2,2)
pylab.plot(y,x,'*g-')
```

```python
# change subplot1 to (2,2,1) and subplot 2 to (2,1,2)

pylab.subplot(2,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(2,1,2)
pylab.plot(y,x,'*g-')
```

```python
# change subplot1 to (2,2,2) and subplot 2 to (2,2,2)

pylab.subplot(2,2,2)
pylab.plot(x,y,'r--')
pylab.subplot(2,2,2)
pylab.plot(y,x,'*g-')
```

```python
""" (1,2,1) and (1,2,2) are significant if we want tilted subplots, without which both the graphs 
will be in the same plot"""

# For example,

pylab.subplot(1,2,1) # The contents of the brackets represents rows,columns and indexes
pylab.plot(x,y,'r--') # The third parameter('r--') determines color and line style
pylab.subplot(1,2,1)
pylab.plot(y,x,'g*-')
```

```python
#Add a Legend on the upper left


import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)

# Create a figure and axes
fig, axes = plt.subplots()  # Corrected to plt.subplots()

# Set labels and title
axes.set_xlabel("X Axis Label")
axes.set_ylabel("Y Axis Label")
axes.set_title("Title of Graph")

# Plotting multiple lines
axes.plot(x, x**2)
axes.plot(x, x**3)

# Adding a legend with specific location
axes.legend(["x = x^2", "x = x^3"], loc='upper left')  # loc=2 can be replaced with loc='upper left'

# Display the plot
plt.show()
```

```python
# Other 2D Graphics

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
n = np.array([0, 1, 2, 3, 4, 5])

# Create subplots
fig, axes = plt.subplots(1, 4, figsize=(16, 5))

# Scatter plot
axes[0].set_title("Scatter")
axes[0].scatter(x, x + 0.25 * np.random.randn(len(x)))

# Step plot
axes[1].set_title("Step")
axes[1].step(n, n**2, linewidth=2)

# Set a title 'step' for next graph (step plot)
axes[1].set_title("step")
axes[1].step(n,n**2,linewidth=2)

plt.show()
```

```python
# Plot a Bar Graph

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
n = np.array([0, 1, 2, 3, 4, 5])

# Create subplots
fig, axes = plt.subplots(1, 4, figsize=(16, 5))

# Scatter plot
axes[0].set_title("Scatter")
axes[0].scatter(x, x + 0.25 * np.random.randn(len(x)))

# Step plot
axes[1].set_title("Step")
axes[1].step(n, n**2, linewidth=2)

# Set a title 'step' for next graph (step plot)
axes[1].set_title("step")
axes[1].step(n,n**2,linewidth=2)

# plot a Bar Graph
axes[2].set_title("bar")
axes[2].bar(n,n**2,align='center', width=0.5, alpha=0.5)

# Plot a fill between graph
axes[3].set_title("fill_between")
axes[3].fill_between(x,x**2,x**3,color='green')
```

```python
# Draw a contour map

import matplotlib.pyplot as plt
import numpy as np

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)

X,Y=np.meshgrid(x,y)

Z1=np.exp(-X**2)-(Y**2)
Z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1-Z2)*2

print(X)
print(Y)
```

```python
# Create a figure

fig, axes=plt.subplots()
CS=axes.contour(X,Y,Z)
axes.clabel(CS,inline=1, fontsize=10)
axes.set_title("Contour Map")
```
